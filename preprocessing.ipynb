```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_images_path = "dataset/images/train"
val_images_path = "dataset/images/val"
train_labels_path = "dataset/labels/train"
val_labels_path = "dataset/labels/val"
```

```python
train_images = os.listdir(train_images_path)
print(len(train_images))
```

```python
val_images = os.listdir(val_images_path)
print(len(val_images))
```

```python
train_labels = os.listdir(train_labels_path)
print(len(train_labels))
```

```python
val_labels = os.listdir(val_labels_path)
print(len(val_labels))
```

```python
from PIL import Image

smallest_width = 10000
smallest_height = 10000
smallest_width_img = ""
smallest_height_img = ""

for img_sample in train_images:
    image = os.path.join(train_images_path, img_sample)
    with Image.open(image) as img:
        width, height = img.size
        if width < smallest_width:
            smallest_width = width
            smallest_width_img = image

        if height < smallest_height:
            smallest_height = height
            smallest_height_img = image

print(f"smallest width: {smallest_width}, smallest_height: {smallest_height}")
print(smallest_width_img)
print(smallest_height_img)
```

```python
smallest_width = 10000
smallest_height = 10000
smallest_width_img = ""
smallest_height_img = ""

for img_sample in val_images:
    image = os.path.join(val_images_path, img_sample)
    with Image.open(image) as img:
        width, height = img.size
        if width < smallest_width:
            smallest_width = width
            smallest_width_img = image

        if height < smallest_height:
            smallest_height = height
            smallest_height_img = image

print(f"smallest width: {smallest_width}, smallest_height: {smallest_height}")
print(smallest_width_img)
print(smallest_height_img)
```

```python
from PIL import Image

width_of_images = []
height_of_images = []

for img_sample in train_images:
    image = os.path.join(train_images_path, img_sample)
    with Image.open(image) as img:
        width, height = img.size
        width_of_images.append(width)
        height_of_images.append(height)
        
fig, ax = plt.subplots()
ax.hist(width_of_images, bins=100, linewidth=0.5, edgecolor="white")
plt.show()
```

```python
val_images_set = set()
val_labels_set = set()

for img in val_images:
    root, ext = os.path.splitext(img)
    val_images_set.add(root)

for label in val_labels:
    root, ext = os.path.splitext(label)
    val_labels_set.add(root)

diff = val_labels_set.difference(val_images_set)
print(diff)
```
